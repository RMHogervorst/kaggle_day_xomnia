```python
import pandas as pd
import numpy as np

```

```python
properties_df = pd.read_csv('./data/properties_2016.csv')
train_df = pd.read_csv('./data/train_2016_v2.csv')
```

```python
feat_df = pd.merge(train_df, properties_df, on='parcelid', how='left')
feat_df.head()
```

```python
pd.options.display.max_rows = 65

dtype_df = feat_df.dtypes.reset_index()
dtype_df.columns = ["Count", "Column Type"]
dtype_df
```

```python
missing_df = feat_df.isnull().sum(axis=0).reset_index()
missing_df.columns = ['column_name', 'missing_count']
missing_df['missing_ratio'] = missing_df['missing_count'] / feat_df.shape[0]
temp = missing_df.ix[missing_df['missing_ratio']>0.6]
temp
```

```python
missing_list = temp['column_name'].tolist()
extra = ['heatingorsystemtypeid','buildingqualitytypeid']
print (missing_list, extra)
```

```python
set(missing_list)&set(extra)
```

```python
feat_df = feat_df.drop(missing_list, axis=1)
```
